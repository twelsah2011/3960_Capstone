```python
# Import necessary libraries
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt

# Define the center of the map and zoom level
center = [53.5444, -113.4909]
zoom = 11

# Define the data for each year
data_2015 = traffic_data_2015[['Latitude', 'Longitude']].values.tolist()
data_2016 = traffic_data_2016[['Latitude', 'Longitude']].values.tolist()
data_2017 = traffic_data_2017[['Latitude', 'Longitude']].values.tolist()
data_2018 = traffic_data_2018[['Latitude', 'Longitude']].values.tolist()

# Create a figure with 2 rows and 2 columns
fig = folium.Figure(width=1000, height=800)
m1 = folium.Map(location=center, zoom_start=zoom)
m2 = folium.Map(location=center, zoom_start=zoom)
m3 = folium.Map(location=center, zoom_start=zoom)
m4 = folium.Map(location=center, zoom_start=zoom)
heatmap1 = HeatMap(data_2015, radius=10)
heatmap2 = HeatMap(data_2016, radius=10)
heatmap3 = HeatMap(data_2017, radius=10)
heatmap4 = HeatMap(data_2018, radius=10)
# Define the locations and tooltips
locations = [    [53.550215, -113.481499],
    [53.500139, -113.526056],
    [53.548139, -113.368194]
]
tooltips = [    'Edmonton Central, 95 Street North of 105 Avenue',    'Edmonton South',    'Edmonton East']

# Add the markers to the maps
for location, tooltip in zip(locations, tooltips):
    folium.Marker(location, tooltip=tooltip).add_to(m1)
    folium.Marker(location, tooltip=tooltip).add_to(m2)
    folium.Marker(location, tooltip=tooltip).add_to(m3)
    folium.Marker(location, tooltip=tooltip).add_to(m4)

heatmap1.add_to(m1)
heatmap2.add_to(m2)
heatmap3.add_to(m3)
heatmap4.add_to(m4)

# Add each map to a subplot in the figure
fig.add_child(m1)
fig.add_child(m2)
fig.add_child(m3)
fig.add_child(m4)

# Set the title for each subplot
m1.get_root().html.add_child(folium.Element("<h3>2015</h3>"))
m2.get_root().html.add_child(folium.Element("<h3>2016</h3>"))
m3.get_root().html.add_child(folium.Element("<h3>2017</h3>"))
m4.get_root().html.add_child(folium.Element("<h3>2018</h3>"))

# Display the figure
display(fig)
```
