```python
df_update = pd.read_csv('updated_dataset.csv')
```

```python
# Load the data

df_update['Date/Time'] = pd.to_datetime(df_update['Date/Time'])
df_update = df_update.drop('PM10 Mass', axis=1)
```

```python
df_update.dropna(inplace=True)
```

```python
df_update.info()
```

```python
# First, we need to import the necessary libraries
import pandas as pd
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.utils.plotting import plot_series
from sktime.forecasting.fbprophet import Prophet
from sktime.forecasting.compose import make_reduction
import warnings
warnings.filterwarnings("ignore")



# Convert the DataFrame into an sktime Series
y = df_update.set_index('Date/Time')['Air Quality Index']

# Split the data into training and testing sets
y_train, y_test = temporal_train_test_split(y, test_size=365)

# Create a Prophet model
prophet = Prophet(yearly_seasonality=True)

# Wrap the Prophet model in an sktime estimator
forecaster = make_reduction(prophet, strategy="recursive", window_length=365)

# Fit the model to the training data
forecaster.fit(y_train)

# Generate predictions for the test data
y_pred = forecaster.predict(fh=[i for i in range(len(y_test))])

# Plot the actual and predicted values
plot_series(y_train, y_test, y_pred, labels=['Training data', 'Testing data', 'Predictions'])

```
